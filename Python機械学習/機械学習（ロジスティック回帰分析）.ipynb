```python
import pandas as pd
import numpy as np
```

```python
df_past = pd.read_csv("./data/user_data.csv")
df_future = pd.read_csv("./data/user_data_future.csv")
```

```python
X_name=['sex', 'student', 'stay time']
X = df_past[X_name]
y = df_past['registration']
```

```python
df_past.columns
```

```python
#import statsmodels.api as sm
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression

#ロジスティック回帰
model = sm.Logit(y,sm.add_constant(X))
result = model.fit()
result.summary()
```

```python
#売上予測
result.predict(sm.add_constant(df_future[X_name]))
```
